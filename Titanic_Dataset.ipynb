### Data Cleaning

```python
# 1. Drop Rows : Heavy Data losses
# 2. Finding mean values of missing columns and fill
# 3. Age, Gender - mean (if-else, Name-Mr,Miss)
resultant_title = [train, test]
for data in resultant_title :

    data['Title'] = data['Name'].str.extract('([A-Za-z]+)\.',expand = False)
```
